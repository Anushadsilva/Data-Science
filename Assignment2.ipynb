## Task 5

```python
# print the number of pixels in the png and jpg images

#number of pixels in png
pix = image_png.shape
rpix = pix[0]
cpix = pix[1]
total_pix = rpix * cpix
print("The PNG image consists of",total_pix,"pixels")

#number of pixels in jpg
jpix = image_jpg.shape
r1pix = jpix[0]
c1pix = jpix[1]
total_pix1 = r1pix * c1pix
print("The JPG image consists of",total_pix1,"pixels")
```

## Task 7

```python
# calculate and print the mean values of each of the 3 colors for the png - over the whole image

mean = image_png.mean(axis=(0,1))

R_mean = mean[0]
G_mean = mean[1]
B_mean = mean[2]

print("Mean of R =",R_mean,"G =",G_mean,"B =",B_mean)
```

## Task 8a

```python
# reverse the png image by flipping horizontally 

reverse_img = image_png[::, ::-1, ::]
plt.imshow(reverse_img)
```

## Task 8b

```python
# perform task 8a above using a numpy function

flippedimage= np.flip(image_png, 1)
plt.imshow(flippedimage)
```

## Task 9

```python
# change the transparency of the lower half of the image

img_png_transp = image_png.copy()

img_png_transp[int(img_png_transp.shape[0]/2):, :, 3] = 0.5
plt.imshow(img_png_transp)
```

## Task 10

```python
# print a negative of the png image

img_png_neg2 = image_png.copy()
img_png_neg2[:, :,[0,1,2]] = 1 - img_png_neg2[:, :, [0,1,2]]
plt.imshow(img_png_neg2)
```

## Task 11

```python
# obtain and print a cropping from the png image

crop = image_png[600:1200, 2:550, :]
plt.imshow(crop)
```

##  Task 12

```python
# print the shape of the png image in grayscale and print the image

image = image_png[:, :, [0,1,2]]
new_image = np.dot(image, np.array([0.21, 0.72, 0.07]))
print("Grayscale shape of the image is", new_image.shape)
plt.imshow(new_image, cmap = "gray")
```

## Task 13

```python
# perform gamma correction on the image and display it

gamma_corr = image_png.copy()
gamma = 1/5
gamma_corr[:, :, [0,1,2]] = gamma_corr[:, :, [0,1,2]]**(gamma)
plt.imshow(gamma_corr)
```

## Task 14

```python
# obtain the img in red, green and blue scale and  horizontally stack/ join them 

rscale = image_png.copy()
rscale[:, :, [1,2]] = 0

gscale = image_png.copy()
gscale[:, :, [0,2]] = 0

bscale = image_png.copy()
bscale[:, :, [0,1]] = 0

join = np.concatenate((rscale, gscale, bscale), axis=1)
plt.imshow(join)
```
